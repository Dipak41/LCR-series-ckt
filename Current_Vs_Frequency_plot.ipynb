###  Plot of graph from the experimental data for LCR series Ckt 

```python
import matplotlib.pyplot as plot
import numpy as np
```

```python
with open ("Experimental_Data.txt", 'r') as D:
    F1 = []    #Create empty list for the X-axis (Frequency). Here 1 for Resistance R1 and 2 for Resistance R2.
    I1 = []    #Create empty list for the Y-axis (Current)
    F2 = []    
    I2 = []
    for line in D:
        #print(line)
        if line.startswith("#"):
            #print(line)
            continue
        else:
            lineparts = line.split(' ')
            #print(lineparts)
            SN = lineparts[0]
            #print(SN)
            R1 = float(lineparts[1])
            #print(R1)
            f1 = float(lineparts[2])
            #print(f1)
            F1.append(f1)
            #print(F1)
            i1 = float(lineparts[3])
            #print(i1)
            I1.append(i1)
            #print(I1)
            R2 = float(lineparts[4])
            #print(R2)
            f2 = float(lineparts[5])
            F2.append(f2)
            #print(F2)
            i2 = float(lineparts[6])
            I2.append(i2)
            #print(I2)
'''Now we are going to plot the Current Vs Frequency plot for two different resistance R1 = 47 Ohm and R2 = 100 Ohm'''
#print(F1,I1,F2,I2)
plot.figure(figsize = [7,6])
plot.plot(F1,I1, color = 'b', marker = 'o', label = 'R1 = 47 $\Omega$')
plot.plot(F2, I2, color = 'r', linestyle = '-.', marker = 'o', markersize = 6, label = 'R2 = 100 $\Omega$')
plot.axis([0, 16, 0, 30])
plot.grid(True)
plot.title('Plot of Current Vs Frequency graph for LCR Ckt', fontsize = 14)
plot.legend(loc = 1, prop = {'size': 14})
plot.xlabel("Frequency (f, kHz)", fontsize = 16, color = "red")
plot.ylabel("Currnet (I, mA)", fontsize = 16, color = "black")
plot.savefig("IvsF.jpg") # We can use different formats of figure while save it.       
```

## Calculation of natural frequency for the oscillator

```python
L = 30e-3   # Value of inductance for the calculation of natural frequency (SI Unit)
C = 0.1e-6  # value of capacitance of capacitor (SI Unit)
f_r = 1/np.pi/2*1/np.sqrt(L*C) # It gives the value in Hz
F_resonant = f_r/1000 # In kHz
print("N. freq. of Oscillator = ", F_resonant, 'kHz' )
    
```
